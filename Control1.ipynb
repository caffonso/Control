```python
# Entrada degrau
t         = np.linspace(0,100,1000)     # variavel tempo: start, stop, num 
u         = np.zeros(len(t))            # inicialização u(t) = 0
u[200:400] = 1                          # u(t-20) = 1

# imprimir grafico
print("tempo",np.round(t[0:5],2),"...",t[999],'\n')
plt.figure()
plt.plot(t,u,'g')
plt.xlabel('Tempo[s]')
plt.title("degrau")
plt.ylabel('u(t)')
plt.show()
```

```python
# Entrada rampa, seno e exponencial
f         = 30                     # frequencia
tau       = 10                     # constante de tempo
k         = 2                      # ganho
u1        = k*t                    # rampa
u2        = np.sin(2*np.pi*f*t)    # seno
u3        = np.exp(-t/tau)         # exponencial

plt.figure(figsize=(16,4))
plt.subplot(1,3,1) 
plt.plot(t,u1,'g')
plt.title("u(t) = rampa")
plt.subplot(1,3,2) 
plt.plot(t,u2,'r')
plt.title('u(t) = sin(t)')
plt.subplot(1,3,3) 
plt.plot(t,u3,'b')
plt.title('u(t) = exp(t)')
plt.show()
```
