## Coordenadas polares

*cartesiano para polar:*

modulo
$$ |a| = \sqrt{a.imag^2 + a.real^2}\  $$

argumento
$$ \theta\ = arctan{a.imag/a.real}   $$

polar
$$ p = |a|< \theta\$$


Polar para cartesiano
$$a = |a|\cos{\theta\}\+j|a|\sin{\theta\}\$$

```python

pol = cmath.polar(a)
print(
  "cartesiano -> polar","\n"  
  "|a|<th = ",pol[0],"<",pol[1],'rad',"\n"
  "|a|<th = ",pol[0],"<",np.rad2deg(pol[1]),'graus',"\n"
  "polar -> cartesiano","\n"    
  " a = ",pol[0]*np.cos(pol[1])+ pol[0]*np.sin(pol[1])*1j,"\n"
)
```

## Calculo matricial complexo

É possivel utilizar o calculo matricial para a resolução de circuitos eletricos. No exemplo a seguir o circuito RLC é resolvido através do método da corrente no ramo:




lab4.png

```python
M0 = np.array([ 
        [7 + 3j,   5j   , 5      ],
        [5j    ,  12 +3j, -(2-2j)],
        [5     , -(2-2j),   17-2j] 
])

v0 = np.array([
    [10],
    [5*np.cos(np.pi/6) +5*np.sin(np.pi/6)*1j],               
    [0]                                   
])


print(
    "M0   =  \n",M0,"\n","\n"
    "M0^2 =  \n",M0**2,"\n","\n"
    "M0.real =  \n",M0.real,"\n","\n"
    "M0.imag =  \n",M0.imag,"\n","\n"
    "M0-1    =   \n",np.linalg.inv(M0),"\n","\n"
     "v0 =",v0
    )

```

*Considere as Matrizes M, I e V, na expressão matricial:*
**$$M.I = v$$**
*É possivel calcular o valorde I através da expressão:*

$$I = M^{-1}.v$$

*o valor de I1 pode ser determinado como a seguir:*

```python
I = np.matmul(np.linalg.inv(M0),v0)
print("i1 = ", np.round(I[0,0],2),"A")
```
